```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# minor ticks
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)
```

```python
def plot(df):
    axes = sns.heatmap(df.sort_index(), square=True, cbar=False)

    # Grid
    axes.xaxis.set_minor_locator(MultipleLocator(1))
    axes.yaxis.set_minor_locator(MultipleLocator(1))
    plt.grid(b=True, axis='x', which='minor', color='#666666', linestyle='-')
    plt.grid(b=True, axis='y', which='minor', color='#666666', linestyle='-')

    # Labels
    axes.tick_params('x', top=True, labeltop=True, labelrotation=60)
```

```python
experiments = pd.read_csv("experiments.csv", index_col="group", dtype={'historical': np.int32, 'rcp26': np.int32, 'rcp45': np.int32, 'rcp85': np.int32})
```

```python
plt.figure(figsize=(16, 45))
plot(experiments)
```
