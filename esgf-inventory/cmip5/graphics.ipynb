```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# minor ticks
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)
```

```python
df = pd.read_csv("cmip5.csv", index_col="master_id", dtype={'pr': np.int32, 'sftlf': np.int32, 'tas': np.int32, 'tasmin': np.int32, 'tasmax': np.int32})
```

```python
fig, ax = plt.subplots(figsize=(1.5,120))
axes = sns.heatmap(df.sort_index(), square=False, cbar=False)

# Grid
axes.xaxis.set_minor_locator(MultipleLocator(1))
axes.yaxis.set_minor_locator(MultipleLocator(1))
plt.grid(b=True, axis='x', which='minor', color='#666666', linestyle='-')
plt.grid(b=True, axis='y', which='minor', color='#666666', linestyle='-')

# Labels
axes.tick_params('x', top=True, labeltop=True, labelrotation=60)
```
