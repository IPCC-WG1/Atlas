```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# minor ticks
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)
```

```python
def plot(df):
    axes = sns.heatmap(df.sort_index(), square=True, cbar=False)

    # Grid
    axes.xaxis.set_minor_locator(MultipleLocator(1))
    axes.yaxis.set_minor_locator(MultipleLocator(1))
    plt.grid(b=True, axis='x', which='minor', color='#666666', linestyle='-')
    plt.grid(b=True, axis='y', which='minor', color='#666666', linestyle='-')

    # Labels
    axes.tick_params('x', top=True, labeltop=True, labelrotation=60)
```

```python
day = pd.read_csv("cmip6-day.csv", index_col="dataset_id", dtype={'pr': np.int32, 'sftlf': np.int32, 'tas': np.int32, 'tasmin': np.int32, 'tasmax': np.int32})
mon = pd.read_csv("cmip6-mon.csv", index_col="dataset_id", dtype={'pr': np.int32, 'sftlf': np.int32, 'tas': np.int32, 'tasmin': np.int32, 'tasmax': np.int32})
```

```python
plt.figure(figsize=(16, 180))

plt.subplot(121, title="Daily")
plot(day)
plt.subplot(122, title="Monthly")
plot(mon)
plt.show()
```
